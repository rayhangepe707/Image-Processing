```python
import numpy as np
from PIL import Image

def manual_convolution2d(image_channel, kernel):
    """
    Fungsi untuk melakukan konvolusi manual pada satu kanal gambar (2D array).
    """
    img_height, img_width = image_channel.shape
    kernel_size = kernel.shape[0]

    # Hitung ketebalan padding (Untuk kernel 5x5, k = 5//2 = 2)
    pad_thickness = kernel_size // 2

    # 1. Tambahkan Zero Padding pada citra input
    padded_image = np.pad(image_channel, pad_width=pad_thickness, mode='constant', constant_values=0)

    # 2. Siapkan matriks kosong untuk menyimpan hasil konvolusi
    output_channel = np.zeros((img_height, img_width))

    # 3. Proses konvolusi (menggeser kernel pada setiap piksel)
    for i in range(img_height):
        for j in range(img_width):
            # Ambil potongan matriks (region) dari citra ber-padding sebesar ukuran kernel
            region = padded_image[i : i + kernel_size, j : j + kernel_size]

            # Kalikan region dengan kernel, lalu jumlahkan seluruh elemennya
            convolution_sum = np.sum(region * kernel)

            # Simpan hasil ke matriks output
            output_channel[i, j] = convolution_sum

    return output_channel

def main():
    # 1. Baca gambar menggunakan PIL lalu konversi ke array Numpy
    img_path = 'input.jpg' # Ganti dengan nama file gambarmu
    img = Image.open(img_path)
    img_array = np.array(img)

    print("Memulai proses konvolusi (ini memakan waktu beberapa detik)...")

    # 2. Definisikan Kernel Rataan 5x5
    kernel_size = 5
    # Matriks 5x5 berisi angka 1, lalu dikalikan 1/25
    kernel_mean = np.ones((kernel_size, kernel_size)) / (kernel_size * kernel_size)

    # 3. Pisahkan kanal warna R, G, B
    R_channel = img_array[:, :, 0]
    G_channel = img_array[:, :, 1]
    B_channel = img_array[:, :, 2]

    # 4. Lakukan konvolusi pada masing-masing kanal
    R_out = manual_convolution2d(R_channel, kernel_mean)
    G_out = manual_convolution2d(G_channel, kernel_mean)
    B_out = manual_convolution2d(B_channel, kernel_mean)

    # 5. Gabungkan kembali ketiga kanal menjadi satu matriks RGB
    # axis=-1 berarti menumpuk di dimensi kedalaman (depth)
    img_output = np.stack((R_out, G_out, B_out), axis=-1)

    # 6. Pastikan nilai piksel tidak melebihi batas 0-255 dan bertipe integer
    img_output = np.clip(img_output, 0, 255).astype(np.uint8)

    # 7. Simpan gambar hasil konvolusi
    final_image = Image.fromarray(img_output)
    final_image.save('output_blur_5x5.jpg')
    print("Konvolusi selesai! Gambar disimpan sebagai 'output_blur_5x5.jpg'")

if __name__ == "__main__":
    main()
```
